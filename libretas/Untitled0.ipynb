# Solución general de la ecuación de difusión.

Sea 

$$\frac{\partial u}{\partial t}=\alpha \frac{\partial^2 u}{\partial x^2}$$

la ecuación de difusión en una dimensión, donde $\alpha$ es una constante, siendo, para el caso actual (difusión de calor), la difusividad y donde además, no se consideran condiciones de frontera ni inicial.

Una solución a la ecuación anterior, en general estará formada por dos variables independientes $t$ y $x$, acompañadas del parámetro $\alpha$. 
A partir de la estructura que se observa en la ecuación, se puede pensar en obtener una solución que sea el producto de funciones de $t$ y $x$ respectívamente. 

Sea 

$$u=f(t)v(x)$$.

Entonces, sustituyendo en la ecuación de difusión se tiene

$$f^{´}(t)v(x)=\alpha f(t)v^{´´}(x)$$.

Luego, para separar las variables, divídase la ecuación anterior por el producto

$$f(t)v(x)$$

obteniendo

$$\frac{f^{´}(t)}{f(t)}=\frac{\alpha v^{´´}(x)}{v(x)}$$.

Una vez en la forma estándar de una ecuación separable, se observa que, puesto que $x$ y $t$ son variables independientes, el único modo en que ambas funciones sean iguales (tal como lo sugiere la expresión anterior), es si cada función resulta ser constante. Por lo que se tienen las siguientes relaciones

$$\frac{f^{´}(t)}{f(t)}=k$$ 

y 

$$\frac{\alpha v^{´´}(x)}{v(x)}=k$$

siendo $k$ una constante arbitraria, la cual, analizando la segunda de las ecuaciones, parece poder tener un valor múltiplo de \alpha, en este caso,$k=\alpha \beta^2$.

La primera de las ecuaciones puede escribirse como

$$\frac{df}{dt}=kf$$

de la cual conocemos, por las técnicas de solución, es satisfecha por 

$$f(t)=c_{1}e^{\alpha \beta^2 t}$$.

Para obtener la solución para la segunda ecuación, considérese la forma que tiene, dada por

$$y^{´´}-a^2y=0$$

la cual se sabe tiene una solución general de la siguiente forma

$$y=c_{2}\cosh{ax}+c_{3}\sinh{ax}$$.

Por lo tanto, la solución a la segunda ecuación es

$$v(x)=c_{2}\cosh{\beta x}+c_{3}\sinh{\beta x}$$.

Puesto que se propuso $u=f(t)v(x)$, se tiene que la solución general a la ecuación de difusión es

$$u=e^{\alpha \beta^2 t}(a\cosh{\beta x}+b\sinh{\beta x})$$

donde $\beta$,$a=c_{1}c_{2}$ y $b=c_{1}c_{3}$, son constantes arbitrarias.

Sin embargo, el proceso anterior considera únicamente un valor positivo para $k$. 

Si $k$ tiene un valor negativo, entonces el proceso es análogo, pero la solución obtenida está dada por

$$u=e^{-\alpha \beta^2 t}(a\cosh{\beta x}+b\sinh{\beta x})$$.

Si $k=0$, entonces la solución es $u=c_{1}+c_{2}x$.


## Solución de la ecuación de difusión con condiciones en la frontera.

El problema con valores en la frontera para la ecuación de difusión es el siguiente

$$\frac{\delta u}{\delta t}=\frac{k\delta^2 u}{\delta x^2} , 0<x<L , t>0$$

$$u(0,t)=0 , u(L,t)=0 , t>0$$

$$u(x,0)=f(x) , 0<x<L$$.

De manera análoga al caso anterior, se propone una separación de variables para poder resolver el problema. Por tanto, se propone una solución del tipo $u(x,t)=f(t)v(x)$ y se llega a

$$\frac{f^{´}(t)}{kf(t)}=\frac{v^{´´}(x)}{v(x)}$$

Sin embargo, nótese que para que la ecuación obtenida se cumpla, se necesita que ambas funciones, $f(t)$ y $v(x)$ sean iguales, para toda $t$, lo cual únicamente ocurre si ambas son iguales a una constante, es decir

$$\frac{f^{´}(t)}{kf(t)}=- \lambda $$

$$\frac{v^{´´}(x)}{v(x)}=- \lambda $$

Luego, aplicando las condiciones de frontera a la solución propuesta

$$u(0,t)=v(0)f(t)=0;u(L,t)=v(L)f(t)$$.

Se tienen dos sistemas de ecuaciones diferenciales ordinarias, a saber

$$f^{´}(t)=- \lambda kf(t) y -v^{´´}(x)\lambda g(x)$$

Iniciando por la segunda ecuación, considérese el hecho de esperar que $f(t)\not=0$ para toda $t$, donde las igualdades en las condiciones se cumplen sólo si $v(0)=0$ y $v(L)=0$. Para poder obtener la solución, al igual que en el caso anterior (solución de la ecuación sin condiciones), debe considerarse a $\lambda$. Cabe mencionar que el problema ahora se transforma en un problema de Sturm-Liouville, 

$$-v^{´´}(x)=\lambda v(x),v(0)=0,v(L)=0$$

el cual permitirá escribir la solución en forma de una serie.

Se toman en cuenta los siguientes casos: $\lambda =0,\lambda <0$ y $\lambda >0$.

Para $\lambda =0$, la solución es $v=c_{1}x+c_{2}$. Las condiciones $v(0)=0$ y $v(L)=0$ aplicadas a esta solución significan que $c_{2}=0$ y $c_{1}=0$. Así, para $\lambda =0$ la única solución del problema es la trivial $v=0$.

Para $\lambda <0$ la ecuación se convierte a una del tipo $y^{´´}-\alpha^2 y=0$, la cual se conoce tiene una solución dada por

$$y=c_{1}\cosh{\alpha x}+c_{2}\sinh{\alpha x}$$

Aplicando las condiciones se llega a 

$v(0)=c_{1}\cosh{0}+c_{2}\sinh{0}=c_{1}1+c_{2}0=c_{1}$

por lo que $v(0)=0$ implica $c_{1}=0$. Luego $v(x)=c_{2}\sinh{\alpha x}$. La segunda condición $v(L)=0$ requiere que $c_{2}\sinh{\alpha L}=0$. Para $\alpha \not= 0$, $\sinh{\alpha L}\not= 0$ y esto implica que $c_{2}=0$. Por tanto, la solución para este caso es nuevamente la trivial.

Para $\lambda >0$, la ecuación es del tipo $y^{´´}+\alpha^2 y=0$, la cual tiene la solución general $y=c_{1}\cos{\alpha x}+c_{2}\sin{\alpha x}$. Luego de aplicar las condiciones se tiene que la solución para este caso está dada por

$$v=c_{2}\sin(n\pi /L)x$$.

Por tanto, la solución al problema queda determinada por la ecuación anterior, en donde $\alpha=n\pi /L$, representa los valores propios al problema de Sturm-Liouville. 

Posteriormente y usando el valor obtenido de \alpha, se consigue la expresión para la solución de la primera ecuación, $f(t)$

$$f(t)=c_{3}e^{-k\alpha^2 t}$$

y sustituyendo en la propuesta de solución se llega a 

$$u=f(t)v(x)=ae^{-k\alpha^2 t}\sin{\frac{n\pi}{L}x}$$

La función producto $u$ satisface las condiciones de frontera, sin embargo el factor $a$ necesita estar en una forma tal que $u$ cumpla también la condición inicial. Es decir, se tendría que elegir un valor para $a$ de tal manera que 

$$u(x,0)=g(x)=a\sin{\frac{n\pi}{L}x}$$

lo cual presenta problemas, pues no lograría ser cierta para cualquier elección arbitraria de $g(x)$ lo que implica que $u(x,t)$ no es una solución del problema en vista de que no cumple la condición inicial. 

Sin embargo, la dificultad anteriormente presentada puede salvarse mediante el principio de superposición. 

Considérese la familia de funciones que satisfacen las condiciones de frontera, es decir, 

$$u_{n}=a_{n}e^{-k\alpha^2 t}\sin{\frac{n\pi}{L}x}$$

Usando superposición

$$u(x,t)=\sum_{n=1}^{\infty}u_{n}$$

se tiene que la expresión anterior debe satisfacer las condiciones de frontera. Luego, aplicando la condición inicial se tiene

$$u(x,0)=g(x)=\sum_{n=1}^{\infty}a_{n}\sin{\frac{n\pi}{L}x}$$

la cual es un desarrollo de semi-intervalo de $g$ en una serie de senos. La cual, al usar la teoría de series de Fourier, permite obtener

$$a_{n}=\frac{2}{L}\int_{0}^{L}g(x)\sin{\frac{n\pi}{L}x}dx$$

Por tanto, la solución al problema de la ecuación de difusión sometida a las condiciones consideradas es

$$u(x,t)=\frac{2}{L}\sum_{n=1}^{\infty}(\int_{0}^{L}g(x)\sin{\frac{n\pi}{L}x}dx)e^{-k\alpha^2 t}\sin{\frac{n\pi}{L}x}$$
